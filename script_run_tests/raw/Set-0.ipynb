# 6. Temperatura

## Cómo llevar a cabo el desafío

En este ejercicio deberas escribir un programa que recibirá un string indicando si se cambia de Celsius a Fahrenheit o viceversa y un número en celsius o fahrenheit según corresponda. Deberás devolver la conversión de la temperatura.

Nota: La fórmula para la conversión es de la forma: $(°F - 32) * \frac{5}{9} = °C$

## Aterrizando el ejercicio

Se te dirá si se convierte de Celsius a Fahrenheit o viceversa, por ejemplo si el string inicial es `celsius` quiere decir que el número que vendrá es una temperatura en Celsius, por lo que debes tranformarlo a Fahrenheit. Lo mismo en caso de que el string inicial sea `fahrenheit` es porque debes convertirlo a Celsius.
Una vez hecha la conversion debes imprimir un texto de la forma:
```
"{°F} grados Fahrenheit equivalen a {°C} grados Celsius." -> En caso de que el string inicial sea 'fahrenheit'
"{°C} grados Celsius equivalen a {°F} grados Fahrenheit." -> En caso contrario
```
Donde debes reemplazar los { } con la temperatura dada y la obtenida respectivamente.

### **¿Cómo es el formato del input ? (Input Format)**
Se te entregará un string indicando el tipo de dato que se te dará, seguido de un float.
### **¿Qué puedo asumir sobre ese input? (Constraints)**
El string siempre será `celsius` o `fahrenheit`.
### **¿Qué debo entregar? (Output Format)**
Un string con el resultado de la conversión de la temperatura según corresponda.

## Ejemplo

**Input Test Case 1:**

```
fahrenheit
50
```

**Output Esperado:**


```
50 grados Fahrenheit equivalen a 10 grados Celsius.
```

**Explicación:**

Si reemplazamos 50 °F en la fórmula anteriormente entregada tendremos que:   
$(50 - 32) * \frac{5}{9} = °C$  
$(18) * \frac{5}{9} = °C$  
$\frac{90}{9} = °C$  
$10 = °C$  
Por lo tanto 50°F son 10°C

**Input Test Case 2:**

```
celsius
50
```

**Output Esperado:**


```
50 grados Celsius equivalen a 122 grados Fahrenheit.
```

**Explicación:**

Si reemplazamos 50 °F en la fórmula anteriormente entregada tendremos que:   
$(F° - 32) * \frac{5}{9} = 50$  
$(F° - 32) * 5 = 450$  
$F° - 32 = 90$  
$F° = 122$  
Por lo tanto 50°C son 122°F
